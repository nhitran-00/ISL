# Lab 3: Linear Regression

```python
import numpy as np
import pandas as pd
from IPython.core.pylabtools import figsize
from matplotlib.pyplot import subplots

import statsmodels.api as sm
from statsmodels.stats.outliers_influence \
    import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

# Simple linear regression

```python
Boston = load_data('Boston')
Boston.columns
```

```python
X = pd.DataFrame({'intercept': np.ones(Boston.shape[0]),
                  'lstat': Boston['lstat']})
X[:4]
```

```python
y = Boston['medv']
model = sm.OLS(y, X)
results = model.fit()
summarize(results)
```

```python
design = MS(['lstat'])
X = design.fit(Boston)
X = design.transform(Boston)
X[:4]
```

```python
design = MS(['lstat'])
X = design.fit_transform(Boston)
X[:4]
```

```python
results.summary()
```

```python
results.params
```

```python
new_df = pd.DataFrame({'lstat': [5,10,15]})
newX = design.transform(new_df)
newX
```

```python
new_predictions = results.get_prediction(newX)
new_predictions.predicted_mean
```

```python
new_predictions.conf_int(alpha=0.05)
```

```python
new_predictions.conf_int(obs=True, alpha=0.05) #obs=True: prediction interval
```

```python
results.params['intercept']
```

```python
fig,ax = subplots(nrows=1, ncols=1)
Boston.plot.scatter(x='lstat', y='medv', ax=ax);

# regression line
x_vals = np.array([Boston.lstat.min(), Boston.lstat.max()])
y_vals = results.params['intercept'] + results.params['lstat'] * x_vals
ax.plot(x_vals, y_vals, 'r--');
# ax.axline(x_vals,y_vals);
```

```python
def abline(ax, b, m, *args, **kwargs):
    """Add a line with slope m and intercept b to ax"""
    xlim = ax.get_xlim()
    ylim = [m* xlim[0] + b, m* xlim[1] + b]
    ax.plot(xlim, ylim, *args, **kwargs)

fig,ax = subplots(nrows=1, ncols=1)

Boston.plot.scatter(x='lstat', y='medv', ax=ax)
abline(ax,
       results.params['intercept'],
       results.params['lstat'],
       'r--',
       linewidth=3)
```

```python
ax = subplots(figsize=(8,8))[1]
ax.scatter(results.fittedvalues, results.resid)
ax.set_xlabel('fitted values')
ax.set_ylabel('residual')
ax.axhline(0, color='k', ls='--');
```

```python
infl = results.get_influence()
ax = subplots(figsize=(8,8))[1]
ax.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag) # leverage
ax.set_xlabel('index')
ax.set_ylabel('leverage')
np.argmax(infl.hat_matrix_diag)
```

# Multiple Linear Regression

```python
X = MS(['lstat', 'age']).fit_transform(Boston)
model1 = sm.OLS(y, X)
results1 = model1.fit()
summarize(results1)
```

```python
terms = Boston.columns.drop('medv')
X = MS(terms).fit_transform(Boston)
model = sm.OLS(y, X)
results = model.fit()
summarize(results)
```

```python
minus_age = Boston.columns.drop(['age', 'medv'])
Xma = MS(minus_age).fit_transform(Boston)
model1 = sm.OLS(y, Xma)
summarize(model1.fit())
```

```python
dir(results)
```

```python
print(results.rsquared)
print(np.sqrt(results.scale))
```

```python
vals = [VIF(X, i)
        for i in range(1,X.shape[1])]

vif = pd.DataFrame({'vif': vals},
                   index=X.columns[1:])
vif
```

```python
X = MS(['lstat',
        'age',
        ('lstat', 'age')]).fit_transform(Boston)
model2 = sm.OLS(y, X)
summarize(model2.fit())
```

```python
X = MS([poly('lstat', degree=2), 'age']).fit_transform(Boston)
model3 = sm.OLS(y, X)
results3 = model3.fit()
summarize(results3)
```

```python
anova_lm(results1, results3)
```

```python
ax = subplots(figsize=(8,8))[1]
ax.scatter(results3.fittedvalues, results3.resid)
ax.set_xlabel('fitted values')
ax.set_ylabel('resid')
ax.axhline(0, color='k', ls='--');
```
