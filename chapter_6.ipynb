## Exercise 5
### 5.a
The ridge regression aims to minimize: $$\sum_{i=1}^{n}{(y_i - \beta_0 - \sum_{j=1}^{p}{\beta_jx_{ij}})^2} + \lambda\sum_{j=1}^{p}{\beta_j^2}$$
Given that $\beta_0 = 0$, $x_{i1} = x_{i2}$, the optimization term becomes: $$\sum_{i=1}^{n}{(y_i - x_i\sum_{j=1}^{p}{\beta_j})^2} + \lambda\sum_{j=1}^{p}{\beta_j^2} (2)$$
Given $n=2$ and $p=2$, the optimization term becomes: $$[y_1 - x_1(\beta_1 + \beta_2)]^2 + [y_2 - x_2(\beta_1 + \beta_2)]^2 + \lambda(\beta_1^2 + \beta_2^2)$$
### 5.b
Since the optimization term aims to minimize RSS through $\beta_j$, to find the minimal value of the optimization term, we take its partial differential equation in each of the $\beta_j$ direction, and then set the partial differential equations to 0. The partial differential equations are:
$$\frac{d}{d\beta_1} = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] + 2\lambda\beta_1$$
$$\frac{d}{d\beta_2} = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] + 2\lambda\beta_2$$
Let $a = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)]$, the 2 partial differentials become:
$$\frac{d}{d\beta_1} = a + 2\lambda\beta_1$$
$$\frac{d}{d\beta_2} = a + 2\lambda\beta_2$$
Setting these 2 partial differential equations to 0, we have the following: $$a + 2\lambda\beta_1 = a + 2\lambda\beta_2 = 0$$
$$\leftrightarrow \beta_1 = \beta_2$$
So, in this setting, the ridge coefficient estimates satisfy: $\hat{\beta_1} = \hat{\beta_2}$
### 5.c
The lasso optimization aims to minimize: $$\sum_{i=1}^{n}{(y_i - \beta_0 - \sum_{j=1}^{p}{\beta_jx_{ij}})^2} + \lambda\sum_{j=1}^{p}{|\beta_j|}$$
Similarly to (a), following the constraints, the optimization term becomes: $$[y_1 - x_1(\beta_1 + \beta_2)]^2 + [y_2 - x_2(\beta_1 + \beta_2)]^2 + \lambda(|\beta_1| + |\beta_2|)$$
### 5.d
Since the optimization term aims to minimize RSS through $\beta_j$, to find the minimal value of the optimization term, we take its partial differential equation in each of the $\beta_j$ direction, and then set the partial differential equations to 0. The partial differential equations are:
$$
\frac{d}{d\beta_1} =
\begin{cases}
-2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] - \lambda & \text{if } \beta_1 < 0 \\
-2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] + \lambda & \text{if } \beta_1 \geq 0
\end{cases}
$$
$$
\frac{d}{d\beta_2} =
\begin{cases}
-2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] - \lambda & \text{if } \beta_2 < 0 \\
-2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] + \lambda & \text{if } \beta_2 \geq 0
\end{cases}
$$
Let $a = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)]$, the 2 partial differentials become:
$$
\frac{d}{d\beta_1} =
\begin{cases}
a - \lambda & \text{if } \beta_1 < 0 \\
a + \lambda & \text{if } \beta_1 \geq 0
\end{cases}
$$
$$
\frac{d}{d\beta_2} =
\begin{cases}
a - \lambda & \text{if } \beta_2 < 0 \\
a + \lambda & \text{if } \beta_2 \geq 0
\end{cases}
$$
Setting these 2 partial differential equations to 0, we have the following cases:
* Case 1: $\beta_1 < 0$ and $\beta_2 \geq 0$, then $a - \lambda = a + \lambda = 0 \leftrightarrow \lambda = 0 \text{ and } a = 0 $. This means that the lasso optimization term is minimal, the model becomes least squares. Then, the coefficient estimates is constrained by: $$a = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] = 0$$ $$\leftrightarrow \beta_1 + \beta_2 = \frac{x_1y_1 + x_2y_2}{x_1^2 + x_2^2}$$
* Case 2: $\beta_1 \geq 0$ and $\beta_2 < 0$, then $a - \lambda = a + \lambda = 0 \leftrightarrow \lambda = 0 \text{ and } a = 0 $. This means that the lasso optimization term is minimal, the model becomes least squares. Similar to case 1, the coefficient estimates is constrained by: $$\beta_1 + \beta_2 = \frac{x_1y_1 + x_2y_2}{x_1^2 + x_2^2}$$
* Case 3: $\beta_1 < 0$ and $\beta_2 < 0$, then the coefficient estimates are constrained by: $$a - \lambda = 0$$ $$\leftrightarrow a = \lambda$$ $$\leftrightarrow a = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] = \lambda$$ $$\leftrightarrow \beta_1 + \beta_2 = \lambda + \frac{x_1y_1 + x_2y_2}{x_1^2 + x_2^2}$$
* Case 4: $\beta_1 \geq 0$ and $\beta_2 \geq 0$, then the coefficient estimates are constrained by: $$a + \lambda = 0$$ $$\leftrightarrow a = -\lambda$$ $$\leftrightarrow a = -2x_1[y_1 - x_1(\beta_1 + \beta_2)] -2x_2[y_1 - x_2(\beta_1 + \beta_2)] = -\lambda$$ $$\leftrightarrow \beta_1 + \beta_2 = -\lambda + \frac{x_1y_1 + x_2y_2}{x_1^2 + x_2^2}$$

Among these 4 cases, only case 3 and case 4 are applicable to the lasso optimization since case 1 and case 2 the model is constrained to become least squares. So, in this setting, the lasso coefficients $\hat{\beta_1} \text{ and } \hat{\beta_2}$ are not unique because the solution is any pair of $\hat{\beta_1} \text{ and } \hat{\beta_2}$ that satisfies the constraints in case 3 and case 4:
$$
\hat{\beta_1} + \hat{\beta_2} =
\begin{cases}
\lambda + \frac{x_1y_1 + x_2y_2}{x_1^2 + x_2^2} & \text{if } \hat{\beta_1} < 0 \text{ and } \hat{\beta_2} < 0 \\
-\lambda + \frac{x_1y_1 + x_2y_2}{x_1^2 + x_2^2} & \text{if } \hat{\beta_1} \geq 0 \text{ and } \hat{\beta_2} \geq 0
\end{cases}
$$

Plotting the optimization term as a function of $\beta_1$, with $\lambda = 3$ and the three values of $y_1 = 2, -2, 1$, we obtain the 3 graphs *Lasso optimization* below. The graphs show that the optimization function has the form of a piecewise linear function. * $y_1 = 1$: $\beta_1 = 0$

This confirms that (6.13) is solve by (6.15):
$$
\hat{\beta}_j^L = \begin{cases} 
y_j - \lambda/2 & \text{if } y_j > \lambda/2; \\
y_j + \lambda/2 & \text{if } y_j < -\lambda/2; \\
0 & \text{if } |y_j| \leq \lambda/2.

\text{ (6.15)}
$$

```python
# 8.a
rng = np.random.default_rng(0)
x_8 = rng.normal(size=100)
epsilon_8 = rng.normal(size=100)

# 8.b
betas_8 = rng.uniform(-10, 10, size=4)
Y_8 = betas_8[0] + betas_8[1]*x_8 + betas_8[2]*x_8**2 + betas_8[3]*x_8**3 + epsilon_8

# 8.c
# generate Dataframe of predictors X, X^2, ... X^10
X_8 = {
    **{'X^{}'.format(i): x_8**i for i in range(1, 11)}
}
X_8 = pd.DataFrame(X_8)

# define the model design object
design = MS(X_8.columns).fit(X_8)

# define the function to calculate negative Cp
def nCp(sigma2, estimator, X, Y):
    """Calculate the negative Cp statistic for a given estimator.
    Parameters
    ----------
    sigma2 : float
        The variance of the error associated with yi.
    estimator : object
        The estimator to use for prediction.
    X : array-like
        The features.
    Y : array-like
        The target variable.
    Returns
    -------
    nCp : float
        The negative Cp statistic."""
    n, d = X.shape
    Yhat = estimator.predict(X)
    RSS = np.sum((Y - Yhat) ** 2)
    return -(RSS + 2*d*sigma2) / n
# calculate sigma2 using the noise in Y
sigma2_8 = np.var(epsilon_8)
# partially copy the nCp function
neg_Cp_8 = partial(nCp, sigma2_8)
```

```python
# 8.c
# define the subset selection strategy object
strategy_fw = Stepwise.first_peak(design,\
                        max_terms=len(X_8.columns),\
                        direction='forward')

# define the model selection object
path_Cp_8c = sklearn_selected(OLS,\
                            strategy=strategy_fw,\
                            scoring=neg_Cp_8)

path_Cp_8c.fit(X_8, Y_8)
model_8c = path_Cp_8c.model_.fit()
print('Using Cp, the model obtained includes the following predictors: {}'.format(path_Cp_8c.selected_state_))
print('The coefficient estimates of the model is:\n{}'.format(model_8c.params))
print('The difference between true beta_j and estimated beta_j is:\n{}'.format(betas_8 - model_8c.params))
```

```python
# 8.d
# define the subset selection strategy object
strategy_bw = Stepwise.first_peak(design,\
                        max_terms=len(X_8.columns),\
                        direction='backwards')

# define the model selection object
path_Cp_8d = sklearn_selected(OLS,\
                            strategy=strategy_bw,\
                            scoring=neg_Cp_8)

path_Cp_8d.fit(X_8, Y_8)
model_8d = path_Cp_8d.model_.fit()
print('Using Cp, the model obtained includes the following predictors: {}'.format(path_Cp_8d.selected_state_))
print('The coefficient estimates of the model is:\n{}'.format(model_8d.params))
print('The difference between true beta_j and estimated beta_j is:\n{}'.format(betas_8 - model_8d.params))
```

```python
# 8.e
scaler = StandardScaler(with_mean=True, with_std=True)

# Set up the cross-validation and lambdas for grid search
K = 5
kfold = skm.KFold(K, shuffle=True, random_state=0)
lambdas_8e = 10 ** np.linspace(8, -2, 50) / Y_8.std()

# Create base Lasso model (ElasticNet with l1_ratio=1)
lasso_8e = skl.ElasticNet(l1_ratio=1, max_iter=10000)

# Create pipeline with scaler and lasso
pipe_8 = Pipeline(steps=[('scaler', scaler), ('lasso', lasso_8e)])

# Set up GridSearchCV with custom scoring function
param_grid_8e = {'lasso__alpha': lambdas_8e}
grid_8e = skm.GridSearchCV(pipe_8,
                        param_grid_8e,
                        cv=kfold,
                        scoring=neg_Cp_8,
                        return_train_score=True)

# Fit the model
grid_8e.fit(X_8, Y_8)
lassoCV_8e = grid_8e.best_estimator_.named_steps['lasso']
coef_8e = pd.Series(np.concatenate(([lassoCV_8e.intercept_], lassoCV_8e.coef_)), index=['intercept', *['beta_{}'.format(i) for i in range(1, len(X_8.columns)+1)]])

# Display results
print(f"Best alpha: {grid_8e.best_params_['lasso__alpha']:.6f}")
print(f"Best score: {-grid_8e.best_score_:.6f}")
print(f"\nBest coefficients:\n{coef_8e}")
```

### 8.f
The results of the the coefficient estimates of the 2 strategies, namely forward stepwise selection and the lasso, show that stepwise selection is better at selecting the true model than the lasso. In particular:
* The forward stepwise selection strategy identifies that the model contains two predictors, namely the intercept and $X^7$. Since it successfully identifies the predictors, it also successfully estimates the predictor coefficients with a relative error of less than 1%, yielding a model that is relatively close to the true model. In particular, at the first iteration, forward stepwise identifies $X^7$ as the predictor with the highest correlation with the response variable $Y$. Once $X^7$ is selected and its coefficient sufficiently estimated, the residual between $|Y_i - \hat{Y_i}|$ is significantly reduced. Therefore, adding any more predictors only increases the residual.
* The lasso strategy, on the other hand, identifies that the model contains several predictors other than $X^7$ and the intercept. Unlike the stepwise selection strategy, the lasso splits the coefficient weights among different predictors to yield the lowest possible $|Y_i - \hat{Y_i}|$. Therefore, it is prone to fitting to noise in the data. Since the lasso is performed on standardized data, it is impossible to directly compare the lasso coefficient estimates with their true values. However, since the lasso also includes noise variables, it shows that the lasso yields a less accurate model compared to the true model.

The result shows that in case of high multicollinearity, the greedy algorithm stepwise selection strategy can yield more accurate results than the lasso since it can ignore the noise variables.
