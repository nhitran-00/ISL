## 4.7.2 Logistic Regression

```python
allvars = Smarket.columns.drop(['Today', 'Direction', 'Year'])
design = MS(allvars)
X = design.fit_transform(Smarket)
y = Smarket.Direction == 'Up'

glm = sm.GLM(y,
             X,
             family=sm.families.Binomial())
results = glm.fit()
summarize(results)
```

```python
results.params
```

```python
results.pvalues
```

```python
probs = results.predict()
probs[:10]
```

```python
labels = np.array(['Down']*1250)
labels[probs>0.5] = 'Up'
labels[:10]
```

```python
confusion_table(labels, Smarket.Direction)
```

```python
# calculate error rate using np.mean()
(507+145)/1250, np.mean(labels == Smarket.Direction)
```

```python
# split data into train and test portions
train = (Smarket.Year < 2005)
Smarket_train = Smarket.loc[train]
Smarket_test = Smarket.loc[~train]
Smarket_test.shape
```

```python
# fit logistic regression model
X_train, X_test = X.loc[train], X.loc[~train]
y_train, y_test = y.loc[train], y.loc[~train]
gml_train = sm.GLM(y_train,
                   X_train,
                    family=sm.families.Binomial())
results = gml_train.fit()
probs = results.predict(exog=X_test)
```

```python
# get labels from data
D = Smarket.Direction
L_train, L_test = D.loc[train], D.loc[~train]
```

```python
# generate confusion table
labels = np.array(['Down']*252)
labels[probs>0.5] = 'Up'
confusion_table(labels, L_test)
```

```python
np.mean(labels == L_test), np.mean(labels != L_test)
```

```python
model = MS(['Lag1', 'Lag2']).fit(Smarket)
X = model.transform(Smarket)
X_train, X_test = X.loc[train], X.loc[~train]
glm_train = sm.GLM(y_train,
                   X_train,
                   family=sm.families.Binomial())
results = glm_train.fit()
probs = results.predict(exog=X_test)
labels = np.array(['Down']*252)
labels[probs>0.5] = 'Up'
confusion_table(labels, L_test)
```

## 4.7.3 Linear Discriminant Analysis

sklearn library standardized procedure:
1. create a classifier without referring to any data
2. fit the classifier with data with the fit() method
3. make predictions using the predict() method
4. if wishing to predict probability, using predict_proba() method

```python
lda = LDA(store_covariance=True)

X_train, X_test = [M.drop(columns='intercept') for M in [X_train, X_test]]
lda.fit(X_train, L_train)
```

```python
# mean of the Normal Distribution of each predictor in each class
lda.means_
```

```python
# prior probability (count of each class over total observation count)
lda.priors_
```

```python
lda.classes_
```

```python
# linear discriminant vector: these values provide the linear combination of Lag1 and Lag2 that are used to form the LDA decision rule. They are the multipliers of the elements of X=x if -0.64 x Lag1 -0.51 x Lag2 is large, then the LDA classifier will predict a market increase, and if it is small, then the LDA classifier will predict a market decline
# i.e. decision boundary in the distribution
lda.scalings_
```

```python
lda_pred = lda.predict(X_test)
confusion_table(lda_pred, L_test)
```

```python
lda_prob = lda.predict_proba(X_test)
np.all(
    np.where(lda_prob[:, 1] >= 0.5, 'Up', 'Down') == lda_pred
)
```

```python
# for more than 2 classes, the labels are chosen as the class whose posterior probability is highest
np.all(
    [lda.classes_[i] for i in np.argmax(lda_prob, axis=1)] == lda_pred
)
```

```python
# use a different threshold value
np.sum(lda_prob[:, 0] >= 0.9)
```

```python
Caravan = load_data('Caravan')
Purchase = Caravan.Purchase
Purchase.value_counts()
```

```python
feature_df = Caravan.drop(columns=['Purchase'])

# standardize the data so that all variables are given a mean 0 and a standard deviation of one so that all variables are on the same scale

scaler = StandardScaler(with_mean=True, with_std=True, copy=True)
scaler.fit(feature_df) # scaling are computed and stored in scaler
X_std = scaler.transform(feature_df) # construct the standardized set of features

# create a Pandas.DataFrame from the standardized data
feature_std = pd.DataFrame(X_std,
                           columns=feature_df.columns)
feature_std.std()
```

```python
# split data into train and test portions
(X_train,
 X_test,
 y_train,
 y_test) = train_test_split(feature_std,
                            Purchase,
                            test_size=1000,
                            random_state=0)
```

```python
knn1 = KNeighborsClassifier(n_neighbors=1)
knn1_pred = knn1.fit(X_train, y_train).predict(X_test)
np.mean(y_test != knn1_pred), np.mean(y_test != 'No')
```

```python
confusion_table(knn1_pred, y_test)
```

```python
# Tuning parameters (i.e. hyperparameters): the number of neighbors in KNN

for K in range(1,6):
    knn = KNeighborsClassifier(n_neighbors=K)
    knn_pred = knn.fit(X_train, y_train).predict(X_test)
    C = confusion_table(knn_pred, y_test)
    temp1 = ('K={0:d}:    # predicted to rent: {1:>2}, ' +
             '  # who did rent {2:d}, accuracy {3: .1%}')
    pred = C.loc['Yes'].sum() # retrieve all 'Yes' predictions, different from the true labels
    did_rent = C.loc['Yes', 'Yes'] # retrieve correct 'Yes' predictions
    print(temp1.format(K,
                       pred,
                       did_rent,
                       did_rent/pred))
```

```python
# logistic regression
logit = LogisticRegression(C=1e10, solver='liblinear')
logit.fit(X_train, y_train)
logit_pred = logit.predict_proba(X_test)
logit_labels = np.where(logit_pred[:, 1] > 5, 'Yes', 'No')
confusion_table(logit_labels, y_test)
```

```python
logit_labels = np.where(logit_pred[:, 1] > 0.25, 'Yes', 'No')
confusion_table(logit_labels, y_test)
```

## 4.7.7 Linear and Poisson Regression on the Bikeshare Data

```python
Bike = load_data('Bikeshare')
Bike.shape, Bike.columns
```

```python
# linear regression
X = MS(['mnth',
        'hr',
        'workingday',
        'temp',
        'weathersit']).fit_transform(Bike)

Y = Bike['bikers']
M_lm = sm.OLS(Y, X).fit()
summarize(M_lm)
```

```python
hr_encode = contrast('hr', 'sum')
mnth_encode = contrast('mnth', 'sum')

X2 = MS([mnth_encode,
        hr_encode,
        'workingday',
        'temp',
        'weathersit']).fit_transform(Bike)

M2_lm = sm.OLS(Y, X2).fit()
S2 = summarize(M2_lm)
S2
```

```python
np.sum((M_lm.fittedvalues - M2_lm.fittedvalues)**2)
```

```python
np.allclose(M_lm.fittedvalues, M2_lm.fittedvalues)
```

```python
coef_month = S2[S2.index.str.contains('mnth')]['coef']
coef_month
```

```python
months = Bike['mnth'].dtype.categories
coef_month = pd.concat([coef_month,
                        pd.Series([-coef_month.sum()], index=['mnth[Dec]'])])
coef_month
```

```python
fig_month, ax_month = subplots(figsize=(8,8))
x_month = np.arange(coef_month.shape[0])
ax_month.plot(x_month, coef_month, marker='o', ms=10)
ax_month.set_xticks(x_month)
ax_month.set_xticklabels([l[5:8] for l in coef_month.index], fontsize=15)
ax_month.set_xlabel('Month', fontsize=20)
ax_month.set_ylabel('Coefficient', fontsize=20);
```

```python
coef_hr = S2[S2.index.str.contains('hr')]['coef']
# coef_hr = coef_hr.reindex(['hr[{0}]'.format(h) for h in range(23)])
coef_hr = pd.concat([coef_hr,
                     pd.Series([-coef_hr.sum()], index=['hr[23]'])])
coef_hr
```

```python
fig_hr, ax_hr = subplots(figsize=(8,8))
x_hr = np.arange(coef_hr.shape[0])
ax_hr.plot(x_hr, coef_hr, marker='o', ms=10)
ax_hr.set_xticks(x_hr[::2])
ax_hr.set_xlabel('Hour', fontsize=20)
ax_hr.set_ylabel('Coefficient', fontsize=20);
```

```python
# Poisson Regression

M_pois = sm.GLM(Y, X2, family=sm.families.Poisson()).fit()
```

```python
# get the coefficients values for mnth and hr
S_pois = summarize(M_pois)
coef_month = S_pois[S_pois.index.str.contains('mnth')]['coef']
coef_month = pd.concat([coef_month,
                        pd.Series([-coef_month.sum()], index=['mnth[Dec]'])])
coef_hr = S_pois[S_pois.index.str.contains('hr')]['coef']
coef_hr = pd.concat([coef_hr,
                     pd.Series([-coef_hr.sum()], index=['hr[23]'])])
```

```python
# plot the coefficients associated with mnth and hr
fig_pois, (ax_month, ax_hr) = subplots(1, 2, figsize=(16,8))

#plot the month coefficients
ax_month.plot(x_month, coef_month, marker='o', ms=10)
ax_month.set_xticks(x_month)
ax_month.set_xticklabels([l[5:8] for l in coef_month.index], fontsize=15)
ax_month.set_xlabel('Month', fontsize=20)
ax_month.set_ylabel('Coefficient', fontsize=20)

#plot the hr coefficients
ax_hr.plot(x_hr, coef_hr, marker='o', ms=10)
ax_hr.set_xticks(x_hr[::2])
ax_hr.set_xlabel('Hour', fontsize=20)
ax_hr.set_ylabel('Coefficient', fontsize=20);
```

```python
# compare the fitted values of the 2 models

fig, ax = subplots(figsize=(8,8))
ax.scatter(M2_lm.fittedvalues,
           M_pois.fittedvalues,
           s=20)
ax.set_xlabel('Linear Regression Fit', fontsize=20)
ax.set_ylabel('Poisson Regression Fit', fontsize=20)
ax.axline([0,0], c='black', linewidth=3, linestyle='--', slope=1);
```
