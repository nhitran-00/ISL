### 1.a
Since $\forall x \leq \xi$, $(x - \xi)_+^3 = 0$. So, when $x \leq \xi$, $f(x) = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3$.

Since $\forall x \leq \xi \text{, } f(x) = f_1(x)$, it follows that: $$\beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 = a_1 + b_1x + c_1x^2 + d_1x^3$$
So, $\forall x \leq \xi \text{, } a_1 = \beta_0$, $b_1 = \beta_1$, $c_1 = \beta_2$, $d_1 = \beta_3$.

### 1.b
Since $\forall x > \xi$, $(x - \xi)_+^3 = (x - \xi)^3 = x^3 - 3x^2\xi + 3x\xi^2 - \xi^3$. So, when $x > \xi$, $$f(x) = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \beta_4(x^3 - 3x^2\xi + 3x\xi^2 - \xi^3)$$.
$$f(x) = (\beta_0 - \beta_4\xi^3) + (\beta_1 + 3\beta_4\xi^2)x + (\beta_2 - 3\beta_4\xi)x^2 + (\beta_3 + \beta_4)x^3$$

Since $\forall x > \xi \text{, } f(x) = f_2(x)$, it follows that: $a_2 = \beta_0 - \beta_4\xi^3$, $b_2 = \beta_1 + 3\beta_4\xi^2$, $c_2 = \beta_2 - 3\beta_4\xi$, $d_2 = \beta_3 + \beta_4$.

### 1.c
Given (a), for $f_1(x)$:
$$f_1(\xi) = a_1 + b_1\xi + c_1\xi^2 + d_1\xi^3$$
$$f_1(\xi) = \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3$$

Given (b), for $f_2(x)$:
$$f_2(\xi) = a_2 + b_2\xi + c_2\xi^2 + d_2\xi^3$$
$$f_2(\xi) = (\beta_0 - \beta_4\xi^3) + (\beta_1 + 3\beta_4\xi^2)\xi + (\beta_2 - 3\beta_4\xi)\xi^2 + (\beta_3 + \beta_4)\xi^3$$
$$f_2(\xi) = \beta_0 + (-\beta_4\xi^3 + 3\beta_4\xi^3 - 3\beta_4\xi^3 + \beta_4\xi^3) + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3$$
$$f_2(\xi) = \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3$$

So, $f_1(\xi) = f_2(\xi)$. So, $f(x)$ is continuous at $\xi$.

### 1.d
For $f_1(x)$:
$$f_1'(x) = \beta_1 + 2\beta_2x + 3\beta_3x^2$$
$$f_1'(\xi) = \beta_1 + 2\beta_2\xi + 3\beta_3\xi^2$$

For $f_2(x)$: 
$$f_2'(x) = (\beta_1 + 3\beta_4\xi^2) + 2(\beta_2 - 3\beta_4\xi)x + 3(\beta_3 + \beta_4)x^2$$
$$f_2'(\xi) = \beta_1 + (3\beta_4\xi^2 - 6\beta_4\xi^2 + 3\beta_4\xi^2) + 2\beta_2\xi + 3\beta_3\xi^2$$
$$f_2'(\xi) = \beta_1 + \beta_2\xi + \beta_3\xi^2$$

So, $f_1'(\xi) = f_2'(\xi)$. So, $f'(x)$ is continuous at $\xi$.

### 1.e

For $f_1(x)$:
$$f_1''(x) = 2\beta_2 + 6\beta_3x$$
$$f_1''(\xi) = 2\beta_2 + 6\beta_3\xi$$

For $f_2(x)$:
$$f_2''(x) = 2(\beta_2 - 3\beta_4\xi) + 6(\beta_3 + \beta_4)x$$
$$f_2''(\xi) = 2(\beta_2 - 3\beta_4\xi) + 6(\beta_3 + \beta_4)\xi$$
$$f_2''(\xi) = 2\beta_2 + (-6\beta_4\xi + 6\beta_4\xi) + 6\beta_3\xi$$
$$f_2''(\xi) = 2\beta_2 + 6\beta_3\xi$$

So, $f_1''(\xi) = f_2''(\xi)$. So, $f''(x)$ is continuous at $\xi$.

From (c), (d), (e), it follows that $f(x)$ is a cubic spline.

## Exercise 2
The objective function for $\hat{g}$ is:
$$ \hat{g} = \arg \min_g \left( \sum (y_i - g(x_i))^2 + \lambda \int [g^{(m)}(x)]^2 dx \right) $$
where $g^{(m)}(x)$ represents the $m\text{th}$ derivatives of $g$.

$\lambda = \infty$ indicates that the penalty term $\lambda \int [g^{(m)}(x)]^2 dx$ dominates. To minimize the objective function, the penalty term should approach 0: $$\lambda \int [g^{(m)}(x)]^2 dx \approx 0$$

This implies that the $m\text{th}$ derivative $g^{(m)}(x) = 0$

### 2.a

For $m = 0$, $g^{(m)}(x) = g(x) = 0$. 

So, the graph of $\hat{g}(x)$ is a horizontal line at $y=0$ 


### 2.b
For $m = 1$, $g^{(1)}(x) = g'(x) = 0$. This implies that $g(x) = c$, where $c$ is the mean of $y$

So, the graph of $\hat{g}(x)$ is a horizontal line at the mean of $y$.

### 2.c
For $m = 2$, $g^{(2)}(x) = g''(x) = 0$. This implies that $g(x) = c_0 + c_1x$. 

So, the graph of $\hat{g}(x)$ is a least squares regression straight line.

### 2.d
For $m = 3$, $g^{(3)}(x) = g'''(x) = 0$. This implies that $g(x) = c_0 + c_1x + c_2x^2$.

So, the graph of $\hat{g}(x)$ is a quadratic curve (parabola).

### 2.f
$\lambda = 0$ implies that the penalty term is canceled from the objective function. So, to minimize the objective function, the error term needs to be minimized. To achieve this, $\hat{g}$ interpolates the values of $y$.

So, the graph of $\hat{g}(x)$ is an curve that passes through every data point.

## Exercise 5
The objective function for $\hat{g}$ is:
$$ \hat{g}_1 = \arg \min_g \left( \sum (y_i - g(x_i))^2 + \lambda \int [g^{(3)}(x)]^2 dx \right) $$
$$ \hat{g}_2 = \arg \min_g \left( \sum (y_i - g(x_i))^2 + \lambda \int [g^{(4)}(x)]^2 dx \right) $$
where $g^{(m)}(x)$ represents the $m\text{th}$ derivatives of $g$.

For both $\hat{g}_1$ and $\hat{g}_2$, the objective function is a $RSS$ plus a penalty term that penalizes the third and fourth derivatives of $g$ respectively.

$\lambda$ is a regularization parameter that controls the smoothness of $g$. The higher the value of $\lambda$, the more the penalty term dominates the objective function, leading to a smoother $g$ as $g$ is forced to change less. Reversely, the lower the value of $\lambda$, the more the $RSS$ dominates the objective function, leading to a more wiggly $g$ as $g$ simply interpolates the data to minimize $RSS$.

### 5.a
As $\lambda \rightarrow \infty$, the penalty term in $\hat{g}_1$ and $\hat{g}_2$ dominates the objective function, forcing the change in $\hat{g}_1$ and $\hat{g}_2$ to be minimal. For the objective function to be minimum:
* For $\hat{g}_1$, $\forall x, g^{(3)}(x) = 0$. This implies that, $\hat{g}_1 = a_1 + b_1x + c_1x^2$. So, $\hat{g}_1$ is a quadratic function.
* For $\hat{g}_2$, $\forall x, g^{(4)}(x) = 0$. This implies that, $\hat{g}_2 = a_2 + b_2x + c_2x^2 + d_2x^3$. So, $\hat{g}_2$ is a cubic function.

Given the polynomial forms of $\hat{g}_1$ and $\hat{g}_2$, since $\hat{g}_1$ has a degree 2 and $\hat{g}_2$ has a degree 3, $\hat{g}_1$ has less flexibility than $\hat{g}_2$, thus $\hat{g}_1$ has higher bias but lower variance than $\hat{g}_2$. So, $\hat{g}_2$ is more likely to fit the training data better than $\hat{g}_1$ as the model has more degrees of freedom, it might overfit to fit the noise in the training data. So, $\hat{g}_2$ is more likely to have a lower training RSS than $\hat{g}_1$.

### 5.b
For unseen test data, whether $\hat{g}_1$ or $\hat{g}_2$ performs better depends on the true relationship between the predictors and response:
* If the true relationship is closer to a quadratic function (e.g linear or quadratic), then $\hat{g}_1$ is more likely to have lower test RSS. As $\hat{g}_2$ has higher flexibility, and thus higher variance, than $\hat{g}_1$, it might overfit to fit the noise in the training data, leading to a poor generalization to unseen data, thus higher test RSS.
* If the true relationship is closer to a cubic function, then $\hat{g}_2$ is more likely to have lower test RSS. As $\hat{g}_1$ has lower flexibility, it might be too rigid, thus having higher bias, to capture the curvature in the data than $\hat{g}_2$. As $\hat{g}_2$ has more degree of freedoms to capture the true relationship in the data, it has lower bias than $\hat{g}_1$. Therefore, as it is able to better capture the relationship in the data, $\hat{g}_2$ is more likely to perform better on test data, thus having lower test RSS.

### 5.c
As $\lambda = 0$, the penalty term in the objective function is minimized, leaving $RSS$ to dominate. The objective function of $\hat{g}_1$ and $\hat{g}_2$ is then equal and is dominated only by the $RSS$.
$$\hat{g}_1 = \hat{g}_2 = \arg \min_g \left( \sum (y_i - g(x_i))^2 \right)$$
In this case, $\hat{g}_1$ and $\hat{g}_2$ are identical to each other. These two models simply interpolates the data to minimize $RSS$. So, their training and test RSS are the same.

### 6.a
Using 5-fold cross-validation, the optimal degree with the lowest cross-validated MSE is $d=4$. This is shown in the plot *CV error by polynomial degree*. The plot shows that from $d=1$ to $d=2$, there is a drastic decrease of 75 units of MSE. From $d=2$ to $d=10$, the fluctutation in CV MSE is within 5 units of MSE. While the minimum CV MSE is at $d=4$, the decrease in CV MSE from $d=3$ to $d=4$ is rather minimal. The CV MSE result implies that a quartic polynomial is the model with the best predictive power, and a cubic polynomial is sufficiently good compared to a quartic polynomial.

Using ANOVA, the optimal degree is $d=3$. This is shown in the *ANOVA table*. The gain in F-value as $d=1$ to $d=2$ is the highest, 143.519 with a significantly low p-value $\left(p << 0.05 \right)$. This implies a significant increase in the model's explanatory power. As $d=2$ to $d=3$, the increase in F-value is still significant, 9.884 with a p-value $\left(p << 0.05 \right)$. This implies that moving from a quadratic to cubic polynomial significantly increases the model's explanatory power. From $d=3$ to $d=10$, the increase in F-value decreases, p-value $\left(p > 0.05 \right)$. Moving from $d=3$ to $d=4$, the p-value is around 0.05111, which borders the threshold 0.05.This implies that moving from a cubic to quartic polynomial doesn't significantly increase the model's explanatory power.

The results show that the optimal degree for model's predictive power is $d=4$ while the optimal degree for model's explanatory power is $d=3$. The two results show that a cubic model is sufficient enough for both predictive and explanatory power, since the gain in predictive power from cubic to quartic is rather minimal.

```python
Wage = load_data('Wage')

X = Wage['age'].values.reshape(-1, 1)
y = Wage['wage'].values.reshape(-1, 1)

degrees = range(1, 11)
cv_scores = []
cv = KFold(n_splits=5,
           shuffle=True,
           random_state=0)
# perform 5 fold cross validation for each degree
for degree in degrees:
    model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('linear', LinearRegression())
    ])
    scores = cross_val_score(model, X, y, cv=cv, scoring='neg_mean_squared_error')
    cv_scores.append(-scores.mean())

# fit models with statsmodels for ANOVA
models = []
for degree in degrees:
    X_poly = PolynomialFeatures(degree=degree).fit_transform(X)
    model = sm.OLS(y, X_poly).fit()
    models.append(model)

# find the optimal degree using cross validation
optimal_degree = degrees[np.argmin(cv_scores)]
print(f"Optimal degree using 5-fold cross validation: {optimal_degree}")

# find the optimal degree using ANOVA
anova_df = anova_lm(*models)
optimal_degree_anova = degrees[anova_df[anova_df['Pr(>F)'] < 0.05].index[-1]]
print(f"Optimal degree using ANOVA: {optimal_degree_anova}")

print(f'ANOVA results:\n{anova_df}')
```

```python
# create visualization
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 12))

# left plot: CV error by degree
ax1.plot(degrees, cv_scores, marker='o', linewidth=2, markersize=8)
ax1.axvline(optimal_degree, color='red', linestyle='--', 
            label=f'Optimal degree = {optimal_degree}')
ax1.set_xlabel('Degree')
ax1.set_ylabel('Cross-validated MSE')
ax1.set_title('CV Error by Polynomial Degree', fontsize=12)
ax1.set_xticks(degrees)
ax1.legend()

# right plot: fitted polynomial with optimal degree
final_model = Pipeline([
    ('poly', PolynomialFeatures(degree=optimal_degree)),
    ('linear', LinearRegression())
])
final_model.fit(X, y)
# use statsmodels OLS to get 95% prediction interval
X_poly = final_model.named_steps['poly'].transform(X)
model_sm = sm.OLS(y, X_poly).fit()
age_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
age_range_poly = final_model.named_steps['poly'].transform(age_range)
preds = model_sm.get_prediction(age_range_poly)
preds_mean = preds.predicted_mean
preds_ci = preds.conf_int(alpha=0.05)

ax2.scatter(X, y, alpha=0.5, s=10, label='Data', color='gray')
for val, fmt in zip([preds_mean, preds_ci[:, 0], preds_ci[:, 1]], ['b', 'r--', 'r--']):
    ax2.plot(age_range, val, fmt, linewidth=2)
ax2.set_xlabel('Age')
ax2.set_ylabel('Wage')
ax2.set_title('Polynomial Regression', fontsize=12)
ax2.legend()
plt.show()
```

### 6.b
Using 5-fold cross validation, the number of knots resulting in the smallest MSE is 10. However, the plot of *CV MSE by number of knots* shows that it is not necessary to use 10 knots. In particular, the plot shows a drastic decrease in CV MSE as the number of knots increases from 1 to 2 (from above 12000 to below 2000). As the number of cuts increases from 2 to 10, the CV MSE decreases minimally. From 6 to 10 cuts, the CV MSE appears to plateau.

Therefore, an optimal number of knots is around 2 to 5.

```python
n_cuts_range = range(1, 11)
cv_scores_step = []

# 5-fold cross validation to find optimal number of cuts
for n_cuts in n_cuts_range:
    age_cuts = pd.qcut(X.reshape(-1), q=n_cuts)
    model = Pipeline([
        ('onehot', OneHotEncoder()),
        ('linear', LinearRegression())
    ])
    scores = cross_val_score(model, age_cuts.reshape(-1, 1), y, cv=cv, scoring='neg_mean_squared_error')
    cv_scores_step.append(-scores.mean())

# find optimal number of cuts
optimal_n_cuts = n_cuts_range[np.argmin(cv_scores_step)]
print(f"Optimal number of cuts: {optimal_n_cuts}")
```

```python
optimal_n_cuts = 5

# fit the step function with the optimal number of cuts
age_bins_final = pd.cut(X.reshape(-1), bins=optimal_n_cuts, labels=False)
X_binned_final = age_bins_final.reshape(-1, 1)
final_model_step = Pipeline([
    ('onehot', OneHotEncoder()),
    ('linear', LinearRegression())
])
final_model_step.fit(X_binned_final, y)

# create prediction visualization
age_range = np.linspace(X.min(), X.max(), 100)
age_bins_pred = pd.cut(age_range, bins=optimal_n_cuts, labels=False)
X_binned_pred = age_bins_pred.reshape(-1, 1)
preds = final_model_step.predict(X_binned_pred)

# create visualization
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 12))

# left plot: CV error by number of cuts
ax1.plot(n_cuts_range, cv_scores_step, marker='o', linewidth=2, markersize=8)
ax1.axvline(optimal_n_cuts, color='red', linestyle='--', 
            label=f'Optimal number of cuts = {optimal_n_cuts}')
ax1.set_xlabel('Number of cuts')
ax1.set_ylabel('Cross-validated MSE')
ax1.set_title('CV Error by Number of Cuts', fontsize=12)
ax1.set_xticks(n_cuts_range)
ax1.legend()

# right plot: fitted step function with optimal number of cuts
ax2.scatter(X, y, alpha=0.5, s=10, label='Data', color='gray')
ax2.plot(age_range, preds, color='blue', linewidth=2, label=f'Step function (n_cuts = {optimal_n_cuts})')
ax2.set_xlabel('Age')
ax2.set_ylabel('Wage')
ax2.set_title('Step Function', fontsize=12)
ax2.legend()
plt.show()
```

## Exercise 7

Explore the relationship between the following variables: *age, maritl, jobclass* to predict *wage*. The model is a GAM with the following form:
$$
wage = \beta_0 + \beta_1f(age) + \beta_2 f(maritl) + \beta_3 f(jobclass)
$$
Since *age* is a continuous variable, we can use a natural spline. Since *maritl* and *jobclass* are categorical, we can use factor terms.


```python
# prepare data
X_age = Wage['age']
X_maritl = Wage['maritl'].cat.codes
X_jobclass = Wage['jobclass'].cat.codes
X = np.column_stack([X_age, X_maritl, X_jobclass])

# fit the GAM model
gam = LinearGAM(s_gam(0) + f_gam(1, lam=0) + f_gam(2, lam=0)).fit(X, y)
```

```python
# visualize the partial effects

fig, axes = plt.subplots(3, 1, figsize=(8, 15))

# age
XX_age = np.linspace(X_age.min(), X_age.max(), 100).reshape(-1, 1)
XX = np.empty((XX_age.shape[0], X.shape[1]))
XX[:, 0] = XX_age.reshape(-1)
XX[:, 1] = X[:, 1].mean(0)
XX[:, 2] = X[:, 2].mean(0)
preds_age, confi_age = gam.partial_dependence(term=0, X=XX, width=0.95)
axes[0].plot(XX_age, preds_age, 'b', linewidth=3, label='Partial Effect of Age')
axes[0].plot(XX_age, confi_age[:, 0], 'r--', linewidth=3, label='95% CI')
axes[0].plot(XX_age, confi_age[:, 1], 'r--', linewidth=3)
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Effect of Age')
axes[0].set_title('Partial dependence of Age on Wage', fontsize=12)
axes[0].legend()

# maritl
maritl_labels = Wage['maritl'].cat.categories
maritl_codes = np.arange(len(maritl_labels))
XX_maritl = gam.generate_X_grid(term=1, n=len(maritl_labels))
XX_maritl[:, 0] = X[:, 0].mean(0)
XX_maritl[:, 2] = X[:, 2].mean(0)
preds_maritl, confi_maritl = gam.partial_dependence(term=1, X=XX_maritl, width=0.95)
yerr_lower = preds_maritl - confi_maritl[:, 0]
yerr_upper = confi_maritl[:, 1] - preds_maritl
axes[1].bar(maritl_codes, preds_maritl, label='Partial Effect of Marital Status')
axes[1].errorbar(maritl_codes, preds_maritl, yerr=[yerr_lower, yerr_upper], fmt='none', capsize=5, label='95% CI', ecolor='black')
axes[1].set_xlabel('Marital Status')
axes[1].set_ylabel('Effect of Marital Status')
axes[1].set_title('Partial dependence of Marital Status on Wage', fontsize=12)
axes[1].set_xticks(maritl_codes)
axes[1].set_xticklabels(maritl_labels, rotation=45, ha='right')
axes[1].legend()

# jobclass
jobclass_labels = Wage['jobclass'].cat.categories
jobclass_codes = np.arange(len(jobclass_labels))
XX_jobclass = gam.generate_X_grid(term=2, n=len(jobclass_labels))
XX_jobclass[:, 0] = X[:, 0].mean(0)
XX_jobclass[:, 1] = X[:, 2].mean(0)
preds_jobclass, confi_jobclass = gam.partial_dependence(term=2, X=XX_jobclass, width=0.95)
yerr_lower = preds_jobclass - confi_jobclass[:, 0]
yerr_upper = confi_jobclass[:, 1] - preds_jobclass
axes[2].bar(jobclass_codes, preds_jobclass, label='Partial Effect of Job Class')
axes[2].errorbar(jobclass_codes, preds_jobclass, yerr=[yerr_lower, yerr_upper], fmt='none', capsize=5, label='95% CI', ecolor='black')
axes[2].set_xlabel('Job Class')
axes[2].set_ylabel('Effect of Job Class')
axes[2].set_title('Partial dependence of Job Class on Wage', fontsize=12)
axes[2].set_xticks(jobclass_codes)
axes[2].set_xticklabels(jobclass_labels, rotation=45, ha='right')
axes[2].legend()

plt.tight_layout()
plt.show()
```
