### 4.a
Since $p = 1$ and $X$ is uniformly (evenly) distributed in the range $[0,1]$, the fraction of the available observations is the ratio between the width of the neighborhood selected for prediction and the range of X.
* The range of X is: $1 - 0 = 1$
* The width of the neighborhood is 10% of the range of X, which is: $0.1 \times 1 = 0.1$
* So, the fraction of the available observations to be used to make prediction is: $0.1 / 1 = 0.1 = 10\%$

### 4.b
Since p = 2 and $X_1$ and $X_2$ are uniformly (evenly) distributed in the range $([0,1], [0,1])$, the fraction of the available observations is the ratio between the area of the neighborhood selected for prediction and the feature space covered by ranges of $X_1$ and $X_2$.
* The feature space size is: $(1 - 0) \times (1 - 0) = 1$
* The area of the neighborhood is 10% of the range of $X_1$ and $X_2$, which is: $(0.1 \times 1) \times (0.1 \times 1) = 0.01$
* So, the fraction of the available observations to be used to make prediction is: $0.01 / 1 = 0.01 = 1\%$

### 4.c
Since p = 100 and $X_i$s are uniformly (evenly) distributed in the range $[0,1]$, the fraction of the available observations is the ratio between the size of the space of the neighborhood selected for prediction and the feature space covered by ranges of $X_i$s.
* The feature space size is: $(1 - 0)^{100} = 1$
* The size of the space of the neighborhood is 10% of the ranges of $X_i$s, which is: $(0.1 \times 1)^{100} = 0.1^{100}$
* So, the fraction of the available observations to be used to make prediction is: $0.01^{100} / 1 = 0.1^{100}$

### 4.d
Parts (a)-(c) show that as p increases, although we still use 10% of the range of each predictor, the neighborhood space selected to make prediction covers an infinitesimally small fraction of the total feature space. So, in high-dimensional spaces, as data points are actually very far apart, there are few training observations that are "near" ay given test observations.

### 4.e
Assume that each measurement $X_i$ of the p-dimensional hypercube follows a uniform distribution.
Let a be the fraction of the range of $X_i$ selected to make the prediction. The length of each side of the hypercube for $p = 1, 2, 100$ are then:
* $p=1$: $a^1 = 0.1 \rightarrow a = 0.1$
* $p=2$: $a^2 = 0.1 \rightarrow a = \sqrt{0.1} = 0.316$
* $p=100$: $a^{100} = 0.1 \rightarrow a = \sqrt[100]{0.1}  = 0.977$

In one dimension, a 10% neighborhood is small. In 100 dimensions, to capture the same 10% of the data, the selected neighborhood must stretch across nearly the entire range (97.7%) of every single feature. The points within this hypercube are not "near" the test point in any meaningful sense; they are far away along most dimensions. As dimensions increase, the volume of the space grows exponentially. To maintain a constant number of data points, the volume of the required neighborhood must also grow exponentially, approaching the volume of the entire feature space. This completely undermines the core assumption of KNN and other local methods. The assumption is that points that are "close" in the feature space have similar responses. But in high dimensions, the points in the selected neighborhood are not truly close. The algorithm is forced to use faraway points to make predictions, which are much less likely to be relevant, severely degrading the model's performance.

$$ P(Y = k \mid X = x)
= \frac{
    \pi_k \,
    \frac{1}{\sqrt{2\pi\sigma_k^2}}
    \exp\left(
      -\frac{(x - \mu_k)^2}{2\sigma_k^2}
    \right)
  }{
    \sum_{j=1}^{K}
    \pi_j \,
    \frac{1}{\sqrt{2\pi\sigma_j^2}}
    \exp\left(
      -\frac{(x - \mu_j)^2}{2\sigma_j^2}
    \right)
  }
$$
